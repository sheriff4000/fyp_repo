```python
import os
import re

# Directory containing the LaTeX table files
tables_dir = 'falm_plots/test'

# Names and letters for the naming convention
names = ["DiversifiedGlobal", "Growth", "Healthcare", "Technology", "Value"]
letters = ["A", "B", "C", "D", "E", "F"]

# Corresponding strategy names
strategies = [
    "FALM1", "RiskyFALM", "GreedyFALM", "SafeFALM", "NeutralFALM", "S&P 500", "Homogenous"
]

def read_table_file(filepath):
    with open(filepath, 'r') as file:
        content = file.read()
    # Extract numbers from the table string
    numbers = re.findall(r"-?\d+\.\d+e?-?\d*?", content)
    # Group numbers into rows of six elements each (excluding the first element in each row which is an index)
    data = [numbers[i:i+6] for i in range(1, len(numbers), 6)]
    return data

# Generate the LaTeX document
# latex_content = """
# \\documentclass{article}
# \\usepackage{graphicx}
# \\usepackage{svg}
# \\usepackage{appendix}
# \\begin{document}
# """
latex_content = """

"""
for name in names:
    for letter in letters:
        filename = f"{name}{letter}_table.tex"
        filepath = os.path.join(tables_dir, filename)
        if os.path.exists(filepath):
            data = read_table_file(filepath)
            table_label = f"tab:{name.lower().replace(' ', '_')}_{letter.lower()}_table"
            table_content = f"""\\begin{{table}}[]
    \\centering
    \\caption{{{name} {letter} Table}}
    \\begin{{tabular}}{{|c|c|c|c|c|c|c|}}
        \\hline
        Strategy & Annual Returns (\\%) & Sharpe & Max \\ac{{DD}} (\\%) & Volatility (\\%)& 5\\% \\ac{{VaR}} (\\%) \\\\
        \\hline
"""
            for strategy, values in zip(strategies, data):
                table_content += f"        {strategy} & {values[0]} & {values[1]} & {values[2]} & {values[3]} & {values[4]} \\\\\n"
            table_content += f"""        \\hline
    \\end{"{tabular}"}
    \\label{{{table_label}}}
\\end{{table}}
"""
            latex_content += table_content + "\n"

latex_content += "\\end{document}"

# Write the LaTeX content to a file
output_file = 'all_tables.tex'
with open(output_file, 'w') as file:
    file.write(latex_content)

print(f"LaTeX document generated: {output_file}")
```
