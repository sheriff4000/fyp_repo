```julia
using Plots

# Generate x values
x = range(0, stop=10, length=100)

# Generate y values for concave growth pattern
y_concave = x.^2

# Generate y values for convex growth pattern
y_convex = sqrt.(x)


# Create the plots
convex = plot(x, y_concave, label="", xticks=[], yticks=[], xlabel="Time", ylabel="Price", title="Convex Growth Pattern", left_margin=5Plots.mm)
concave = plot(x, y_convex, label="", xticks=[], yticks=[], xlabel="Time", ylabel="Price", title="Concave Growth Pattern", left_margin=5Plots.mm)
display(concave)
display(convex)
savefig(concave, "falm_plots/concave.svg")
savefig(convex, "falm_plots/convex.svg")
```

```julia
using Plots
using Statistics
using RollingFunctions

# Generate sample data
prices = [100, 105, 110, 115, 120, 115, 110, 105, 100, 95, 90, 85, 80]
prices = [prices;reverse(prices)]

# Calculate moving averages
short_ma = runmean(prices, 2)
long_ma = runmean(prices, 4)
# xl = 5:length(prices)


# Create the plot
p = plot(prices, label="Prices", marker=:circle, markersize=4)
plot!(p, short_ma, label="Short MA", linewidth=2)
plot!(p, long_ma, label="Long MA", linewidth=2)

# Add markers for buy and sell points
buy_points = prices .< short_ma
sell_points = prices .> short_ma
scatter!(p, findall(buy_points), prices[buy_points], label="Sell", markersize=6, markercolor=:red)
scatter!(p, findall(sell_points), prices[sell_points], label="Buy", markersize=6, markercolor=:green)

xlabel!("Time")
ylabel!("Price")
title!("Moving Average Strategy")
display(p)
savefig(p, "moving_average.svg")
# legend!(:topright)
```
